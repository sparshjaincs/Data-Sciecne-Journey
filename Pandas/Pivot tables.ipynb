```python
#Pivot tables is a special  kind of datagrame used for summarizing data for a 
# particular purpose.

# Rows in pivot table represents variable that we interested in 
# and Column another, and the cell's some aggregate value.

# A Pivot tables also tends to include marginal value as well which are sums of 
# each columns and row. This allows you to able to see the relationship between 
# two variables as just as glance.
```

```python
# Let's import pandas as normal

import pandas as pd
import numpy as np



 
```

```python
df = pd.read_csv("../datasets/cwurData.csv")
df.head()
```

```python
def create(index):
    if index["world_rank"] >=1 and index["world_rank"] <=100:
        return "First Tier Top University"
    elif index["world_rank"] >=101 and index["world_rank"] <=200:
        return "Second Tier Top University"
    elif index["world_rank"] >=201 and index["world_rank"] <=300:
        return "Third Tier Top University"
    else:
        return "Other Top University"
    
df["Rank_Level"] =  df.apply(create, axis=1)
```

```python
df.head()
```

```python
df.pivot_table(values='score',index="country",columns="Rank_Level", aggfunc=[np.mean]).head()
```

```python
df.pivot_table(values='score',index="country",columns="Rank_Level", aggfunc=[np.mean, np.max]).head()
```

```python
df.pivot_table(values='score',index="country",columns="Rank_Level", aggfunc=[np.mean],
              margins = True).head()
```

```python
new_df = df.pivot_table(values='score',index="country",columns="Rank_Level", aggfunc=[np.mean, np.max],
              margins = True)
```
