```python
# in this lecture we're going to take a look on hypothesis testing.
# The goal of hypothesis testing is to determine if the two different conditions 
# we have in an experiement have resulted in different impacts.


import numpy as np
import pandas as pd 
from scipy import stats



```

```python
# Now let's import grades.csv
df = pd.read_csv("../Datasets/grades.csv")
df.head()
```

```python
# When we do hypothesis testing we have two ststements of intrests.
# first one is our actual explanation : alternative hypothesis
# second one is explantion that is not sufficient : null hypothesis

# In hypothesis testing our aim tis to determine whether null hypothesis is true or not
# If we have find that there is a difference between groups, then we can reject the null hypotheis
# and we accept the actual hypothesis

```

```python
print(f"There are {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
# lets create two diffrent experiments by dividing the whole dataframe into two parts.
# first part contains all students who completed course before the end december 2015 : early finishers
# second part will be rest : late finishers

early_finishers = df[pd.to_datetime(df["assignment1_submission"]) < "2016"]
early_finishers.head()
```

```python
late_finishers = df[~df.index.isin(early_finishers.index)]
```

```python
late_finishers
```

```python
# Calculating the mean for both early and late finishers

print(early_finishers["assignment1_grade"].mean())
print(late_finishers["assignment1_grade"].mean())
```

```python
# Ok this looks pretty similar But is it really?
# This is where the students t test comes in play.
# It allows us to form alternative hypothesis("This is different") as well as
# null Hypothesis ("These are same") and then test the null hypothesis

# For doing hypothesis trsting we need to slect the threshold level(alpha)
# till how much chance we are willing to accept. So for this case let's use a thershold 
# 0.05 for our alpha or 5%

# In Scipy library we have ttest_ind() function that performs an independent ttest
# (Population are not related to one another). The result of ttest are t-statistics
# and p-value. the seconf one p-value is what we are interested in
# it indicates the chance of null hypothesis being true.
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(early_finishers["assignment1_grade"], late_finishers["assignment1_grade"])
```

```python
# So we can see that p-value is greater than 5% so the chance of similarity is grated than our alpha
# that's why we cant reject the null hypothesis. 
print(ttest_ind(early_finishers["assignment2_grade"], late_finishers["assignment2_grade"]))
print(ttest_ind(early_finishers["assignment3_grade"], late_finishers["assignment3_grade"]))
print(ttest_ind(early_finishers["assignment4_grade"], late_finishers["assignment4_grade"]))
print(ttest_ind(early_finishers["assignment5_grade"], late_finishers["assignment5_grade"]))
print(ttest_ind(early_finishers["assignment6_grade"], late_finishers["assignment6_grade"]))
```

```python
# By looking the above data one thing that we can conclude is the dataset doesn'y
# enough certanity that can make null hypothesis rejectable.

# There are other techniques that are being used more regularly.
# confidence intervals and bayesian analyses



```

```python
df1 = pd.DataFrame(np.random.random(100) for i in range(100))
df2 = pd.DataFrame(np.random.random(100) for i in range(100))

df1.head()
```

```python
def test(alpha = 0.05):
    num_diff = 0
    for col in df1.columns:
        testval, pval = ttest_ind(df1[col], df2[col])
        if pval <= alpha:
            print(f"Col {col} is statistically significantly different at alpha = {alpha}, pval = {pval}")
            num_diff += 1 
    print(f"Total number different was {num_diff}, which is {float(num_diff) / float(len(df1.columns)) * 100.0} ")
test()
```
