```python
# We all have read and learnt set thoery in mathematics and also have learnt 
# joins in sql Right?

# Well these both concepts are very important while working on data analysis
# and data manipulation. These concepts let you to merge data from different sources
# on the basis of verious conditions which i m sure you already have
# learnt.

# These coditions are : 
# 1. A U B (Full Outer Join)
# 2. A Intersection B (Inner Join)
# 3. A - B (left Inner Join)
# 4. B - A (right Inner Join)

# Consider A as left side data frame and B as Right side dataframe

# You might be wondering how we can implement these concepts in data manipulation
# So In pandas we have a function "Merge()" that let's you  merge two data frame
# on the basis of conditions you want

# Let's import pandas first

import pandas as pd
```

```python
# Creating two dataframes left and right

left = pd.DataFrame([
    {"Name" : "Kelly", "Role" : "Director of HR"},
    {"Name" : "Sally", "Role" : "Course liasion"},
    {"Name" : "James", "Role" : "Grader"}
]) 
left.set_index("Name",inplace = True)
right = pd.DataFrame([
    {"Name" : "Mike", "School" : "Law"},
    {"Name" : "Sally", "School" : "Engineering"},
    {"Name" : "James", "School" : "Business"}
]) 
right.set_index("Name",inplace = True)

# all right we have left and right dataframe and you can see that
# some of index have both role and school assigned so why not we create a new df
# that have three columns Name , Role and School and have the data according to left and right dataframes

# So if we want union of these dataframes, we would call a merge () function and
# passing the left and right df and telling merge that we want to use outer join


```

```python
outer = pd.merge(left, right, how="outer",   left_index = True, right_index = True)

outer

# Union , Outer Join 
```

```python
# if we want to get the intersection we just need to set how to inner
# BY default value of how is inner

inner = pd.merge(left, right, how= "inner", left_index = True, right_index = True)

inner
```

```python
# Similarly we can also perform left inner join and right inner join

left_join = pd.merge(left, right, how="left", left_index = True, right_index = True)
left_join
```

```python
right_join = pd.merge(left, right, how="right", left_index = True, right_index = True)
right_join
```

```python
# we can also do it by another way. there is another parameter in merge function
# "on" that specifies in which column we need to apply join on.

# for example

left = left.reset_index()
right = right.reset_index()

pd.merge(left, right, how="right", on="Name")
```
